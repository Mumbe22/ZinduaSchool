ALIASING

```python
%%sql
SELECT customers.FirstName AS "customer_name",
customers.LastName AS "customer_surname",
employees.FirstName AS "employee_name",
employees.LastName AS "employee_surname"
FROM customers, employees
WHERE employees.Country="Canada" AND customers.SupportRepId=employees.EmployeeId
```

JOINS

1. Inner join - common info/relatiobships in between tables
2. Right/left join - common info based on one table
3. Cross joins - combinations of rows in tables

```python
%%sql
SELECT a.Title AS "album_name",
       a.AlbumID AS "album_id",
       t.Name AS "track_name"
FROM albums a
INNER JOIN tracks t
ON a.Title=t.Name
LIMIT 15;
```

```python
%%sql
SELECT a.Title AS "album_name",
       a.AlbumID AS "album_id",
       t.Name AS "track_name",
       ar.Name AS "artist_name"
FROM albums a
INNER JOIN tracks t
ON a.Title=t.Name
INNER JOIN artists ar
ON a.ArtistId=ar.ArtistId
LIMIT 15;
```
